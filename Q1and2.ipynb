```python
# convert the date columns to datetime format
sikka_user_sign_up['signed_up_on'] = pd.to_datetime(sikka_user_sign_up['signed_up_on'])
sikka_user_sign_up['last_login_date'] = pd.to_datetime(sikka_user_sign_up['last_login_date'])

# calculate the difference between the date columns in days
sikka_user_sign_up['duration'] = (sikka_user_sign_up['last_login_date'] - sikka_user_sign_up['signed_up_on']).dt.days
```

```python
pro_sikka_user_sign_up = df_Q2_users_signup.loc[df_Q2_users_signup['app_id'] == 'sikka_pro']
```

```python
# convert the date columns to datetime format
pro_sikka_user_sign_up['signed_up_on'] = pd.to_datetime(pro_sikka_user_sign_up['signed_up_on'])
pro_sikka_user_sign_up['last_login_date'] = pd.to_datetime(pro_sikka_user_sign_up['last_login_date'])

# calculate the difference between the date columns in days
pro_sikka_user_sign_up['duration'] = (pro_sikka_user_sign_up['last_login_date'] - pro_sikka_user_sign_up['signed_up_on']).dt.days

pro_sikka_user_sign_up.head()
```

```python
sikka_left_join_df = pd.merge(df_Q2_rewards_details, sikka_User_offer_completion_data, on='reward_id', how='left')
```

```python
sikka_left_join_df
```

```python
sikka_left_join_df = pd.merge(sikka_left_join_df, sikka_user_sign_up, on='user_id', how='left')
```

```python
sikka_left_join_df.info()
```

```python
sikka_new_df = sikka_left_join_df.groupby('user_id').sum()[['total_payout_in_paise','total_revenue_in_paise','duration']]
```

```python
sikka_new_df.head(50)
```

```python
payout_sikka = sikka_new_df['total_payout_in_paise'].sum()
payout_sikka
```

```python
Revenue_sikka = sikka_new_df['total_revenue_in_paise'].sum()
Revenue_sikka
```

```python
sikka_new_df['duration'].sum()
```

```python
Average_life_span_users_sikka = sikka_new_df['duration'].sum() / len(sikka_new_df)
Average_life_span_users_sikka
```

```python
Net_Revenue_sikka = Revenue_sikka - payout_sikka
Net_Revenue_sikka
```

```python
Average_Revenue_per_user = Net_Revenue_sikka / (len(sikka_new_df))
Average_Revenue_per_user
```

```python
CustomerLifeValue_sikka = (Average_Revenue_per_user) * (Average_life_span_users_sikka)
CustomerLifeValue_sikka
```
